```python
# name 이용
data['NEW_Initial']= train.Name.str.extract('([A-Za-z]+)\.') #lets extract the Salutations
# train['NEW_Initial']= train.Name.str.extract('([A-Za-z]+)\.') #lets extract the Salutations
# test['NEW_Initial']= test.Name.str.extract('([A-Za-z]+)\.') #lets extract the Salutations
```

```python
data['NEW_Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don', 'Dona'],
                        ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr', 'Mr'],inplace=True)
# train['NEW_Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don', 'Dona'],
#                          ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr', 'Mr'],inplace=True)
# test['NEW_Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don', 'Dona'],
#                          ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr', 'Mr'],inplace=True)
```

```python
data['NEW_Master'] = data['NEW_Initial'].map({'Master':1, 'Miss':0, 'Mr':0, 'Mrs':0, 'Other':0})
data['NEW_Miss'] = data['NEW_Initial'].map({'Master':0, 'Miss':1, 'Mr':0, 'Mrs':0, 'Other':0})
data['NEW_Mr'] = data['NEW_Initial'].map({'Master':0, 'Miss':0, 'Mr':1, 'Mrs':0, 'Other':0})
data['NEW_Mrs'] = data['NEW_Initial'].map({'Master':0, 'Miss':0, 'Mr':0, 'Mrs':1, 'Other':0})
```

```python
# Age NA 데이터 분석
#train[pd.isna(train['Age'])]
train[pd.isna(train['Age'])].sort_values(by=['Pclass'])
```

```python
Master_age = data.groupby('NEW_Initial').mean()['Age']['Master']
Miss_age = data.groupby('NEW_Initial').mean()['Age']['Miss']
Mr_age = data.groupby('NEW_Initial').mean()['Age']['Mr']
Mrs_age = data.groupby('NEW_Initial').mean()['Age']['Mrs']
Other_age = data.groupby('NEW_Initial').mean()['Age']['Other']
```

```python
# Initial로 age 채우기
data['NEW_Age'] = data['Age']
data.loc[(data.Age.isnull())&(data.NEW_Initial=='Master'),'NEW_Age'] = Master_age
data.loc[(data.Age.isnull())&(data.NEW_Initial=='Miss'),'NEW_Age'] = Miss_age
data.loc[(data.Age.isnull())&(data.NEW_Initial=='Mr'),'NEW_Age'] = Mr_age
data.loc[(data.Age.isnull())&(data.NEW_Initial=='Mrs'),'NEW_Age'] = Mrs_age
data.loc[(data.Age.isnull())&(data.NEW_Initial=='Other'),'NEW_Age'] = Other_age
```

```python
data['NEW_Age']
```

```python
# Age NA 채우기
## case1 : 범주형으로 분할
def age_split(x):
    if x < 20:
        return 10
    elif x < 40:
        return 2030
    elif x < 60:
        return 4050
    elif x < 80:
        return 6070
    else:
        return np.nan

data['NEW_Age2'] = data['NEW_Age'].map(lambda x: age_split(x))
# train['NEW_Age'] = train['Age'].map(lambda x: age_split(x))
# test['NEW_Age'] = test['Age'].map(lambda x: age_split(x))
# age_010 = len(train[train['NEW_Age']==10])/(pd.isna(train['NEW_Age'])==False).sum()
# age_2030 = len(train[train['NEW_Age']==2030])/(pd.isna(train['NEW_Age'])==False).sum()
# age_4050 = len(train[train['NEW_Age']==4050])/(pd.isna(train['NEW_Age'])==False).sum()
# age_6070 = len(train[train['NEW_Age']==6070])/(pd.isna(train['NEW_Age'])==False).sum()
# train['NEW_Age'] = train['NEW_Age'].fillna(np.random.choice([10,2030,4050,6070], p =[age_010, age_2030, age_4050, age_6070]))
# test['NEW_Age'] = test['NEW_Age'].fillna(np.random.choice([10,2030,4050,6070], p =[age_010, age_2030, age_4050, age_6070]))

data['NEW_Age_010'] = data['NEW_Age2'].map({10:1, 2030:0, 4050:0, 6070:0})
data['NEW_Age_2030'] = data['NEW_Age2'].map({10:0, 2030:1, 4050:0, 6070:0})
data['NEW_Age_4050'] = data['NEW_Age2'].map({10:0, 2030:0, 4050:1, 6070:0})
data['NEW_Age_6070'] = data['NEW_Age2'].map({10:0, 2030:0, 4050:0, 6070:1})
```

```python
# Age NA를 Lasso로 해보겠다는 강한 포부 - FAIL
data_age = data[['Age', 'SibSp', 'Parch', 'NEW_Cabin', 'NEW_Master', 'NEW_Miss', 'NEW_Mr', 'NEW_Mrs', 'NEW_Other', 'NEW_Sex',
                 'NEW_Pclass_1', 'NEW_Pclass_2', 'NEW_Pclass_3', 'NEW_Embarked_C', 'NEW_Embarked_Q', 'NEW_Embarked_S', 'NEW_Fare']]
data_age_train_X = data_age[-pd.isna(data['Age'])]
data_age_test_X = data_age[pd.isna(data['Age'])]
data_age_train_Y = data_age_train_X['Age']
del data_age_train_X['Age']
del data_age_test_X['Age']

from sklearn.linear_model import Lasso
age_lasso = Lasso(alpha=1.0, fit_intercept=True, normalize=False, precompute=False, copy_X=True, max_iter=1000, 
                    tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')
age_lasso.fit(data_age_train_X, data_age_train_Y)
data_age_test_Y = age_lasso.predict(data_age_test_X)

data['NEW_Age'] = data_age['Age'].fillna(pd.Series(data_age_test_Y))
```

```python
# Age one-hot encoding
def age_split(x):
    if x < 20:
        return 10
    elif x < 40:
        return 2030
    elif x < 60:
        return 4050
    elif x <= 80:
        return 6070
    else:
        return np.nan
data['NEW_Age2'] = data['NEW_Age'].map(lambda x: age_split(x))

data['NEW_Age_010'] = data['NEW_Age2'].map({10:1, 2030:0, 4050:0, 6070:0})
data['NEW_Age_2030'] = data['NEW_Age2'].map({10:0, 2030:1, 4050:0, 6070:0})
data['NEW_Age_4050'] = data['NEW_Age2'].map({10:0, 2030:0, 4050:1, 6070:0})
data['NEW_Age_6070'] = data['NEW_Age2'].map({10:0, 2030:0, 4050:0, 6070:1})
```

모델링(stacking by randomForest, bagging, boosting) - validation

```python
# train - validation split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
train_val_X, validation_X, train_val_Y, validation_Y = train_test_split(train_X, train_Y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

model_rf = RandomForestClassifier(n_estimators=200, max_depth=10, min_samples_split=2, min_samples_leaf=1, 
                                  min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, 
                                  min_impurity_decrease=0.0, min_impurity_split=None, bootstrap=True, 
                                  oob_score=False, random_state=None, verbose=0, warm_start=False, class_weight=None)

model_bag = BaggingClassifier(base_estimator=None, n_estimators=200, bootstrap=True, 
                              bootstrap_features=False, oob_score=False, warm_start=False, random_state=None, verbose=0)

model_boost = GradientBoostingClassifier(loss='deviance', learning_rate=0.401, n_estimators=100, subsample=1.0, 
                                         criterion='friedman_mse', min_samples_split=2, min_samples_leaf=1, 
                                         min_weight_fraction_leaf=0.0, max_depth=3, min_impurity_decrease=0.0, 
                                         min_impurity_split=None, init=None, random_state=None, max_features=None, 
                                         verbose=0, max_leaf_nodes=None, warm_start=False, presort='auto')

model_knn = KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, 
                                 p=2, metric='minkowski', metric_params=None)

model_logi = LogisticRegression(penalty='l2', dual=False, tol=0.01, C=0.3, fit_intercept=True, intercept_scaling=1, class_weight=None, 
                                random_state=None, max_iter=100, verbose=0, warm_start=False)

model_xgb = XGBClassifier(max_depth=3, learning_rate=0.9, n_estimators=500, silent=True, objective='binary:logistic', booster='gbtree', 
                          n_jobs=1, nthread=None, gamma=0, min_child_weight=1, max_delta_step=0, subsample=1, colsample_bytree=1, 
                          colsample_bylevel=1, reg_alpha=0, reg_lambda=1, scale_pos_weight=1, base_score=0.5, random_state=0, seed=None, missing=None)

model_all = VotingClassifier(estimators=[('rf', model_rf), ('bag', model_bag), ('boost', model_boost), 
                                         ('knn', model_knn), ('logi', model_logi), ('xgb', model_xgb)])

model_all.fit(train_val_X, train_val_Y)
pred_val_all = model_all.predict(validation_X)
accuracy_score(validation_Y, pred_val_all)
```

```python
# randomforest
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=200, max_depth=5, min_samples_split=2, min_samples_leaf=1, 
                                  min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, 
                                  min_impurity_decrease=0.0, min_impurity_split=None, bootstrap=True, 
                                  oob_score=False, random_state=None, verbose=0, warm_start=False, class_weight=None)
model_rf.fit(train_val_X, train_val_Y)
pred_val_rf = model_rf.predict(validation_X)
accuracy_score(validation_Y, pred_val_rf)
```

```python
# bagging
from sklearn.ensemble import BaggingClassifier
model_bag = BaggingClassifier(base_estimator=None, n_estimators=200, bootstrap=True, 
                              bootstrap_features=False, oob_score=False, warm_start=False, random_state=None, verbose=0)
model_bag.fit(train_val_X, train_val_Y)
pred_val_bag = model_bag.predict(validation_X)
accuracy_score(validation_Y, pred_val_bag)
```

```python
# boosting
from sklearn.ensemble import GradientBoostingClassifier
model_boost = GradientBoostingClassifier(loss='deviance', learning_rate=0.2, n_estimators=100, subsample=1.0, 
                                         criterion='friedman_mse', min_samples_split=2, min_samples_leaf=1, 
                                         min_weight_fraction_leaf=0.0, max_depth=3, min_impurity_decrease=0.0, 
                                         min_impurity_split=None, init=None, random_state=None, max_features=None, 
                                         verbose=0, max_leaf_nodes=None, warm_start=False, presort='auto')
model_boost.fit(train_val_X, train_val_Y)
pred_val_boost = model_boost.predict(validation_X)
accuracy_score(validation_Y, pred_val_boost)
```

```python
model_logi = LogisticRegression(penalty='l2', dual=False, tol=0.01, C=0.3, fit_intercept=True, intercept_scaling=1, class_weight=None, 
                                random_state=None, max_iter=100, verbose=0, warm_start=False)
model_logi.fit(train_val_X, train_val_Y)
pred_val_logi = model_logi.predict(validation_X)
accuracy_score(validation_Y, pred_val_logi)
```
